```python
from tensorflow.keras import datasets
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
(train_images,train_labels),(test_images,test_labels) = datasets.cifar10.load_data()

```

```python
len(train_images)
len(test_images)
train_images[0].shape
train_images[0]
train_labels
train_images,test_images=train_images/255.0,test_images/255.0
class_names=['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
IMG_INDEX = 7
plt.imshow(train_images[IMG_INDEX])
plt.xlabel(class_names[train_labels[IMG_INDEX][0]])
plt.show()
```

```python
train_images.shape
test_images.shape
n_samples, n_x, n_y, n_rgb = train_images.shape
```

```python
train_images = train_images.reshape((n_samples,n_x*n_y*n_rgb))
train_images.shape
n_samples,n_x,n_y,n_rgb = test_images.shape
rss = n_samples,n_x*n_y*n_rgb
test_images = test_images.reshape(rss)
test_images.shape
model = DecisionTreeClassifier(criterion = "gini", max_depth=None)
```

```python
model = DecisionTreeClassifier(criterion="entropy", max_depth=3)
model.fit(train_images, train_labels)

predictions = model.predict(test_images)
cm = confusion_matrix(test_labels, predictions)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot()
plt.show()
```
